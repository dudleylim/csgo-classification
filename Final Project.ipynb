```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
pd.set_option('display.max_columns', None)
```

```python
url = 'datasets/csgo_round_snapshots.csv'
df = pd.read_csv(url)
df.head(25)
```

# Data Preprocessing

```python
# replace categorical with numerical representative
df['round_winner'].replace(['T', 'CT'], [0, 1], inplace=True)
```

```python
df['map'].replace(['de_dust2', 'de_mirage', 'de_nuke', 'de_inferno', 'de_overpass',
       'de_vertigo', 'de_train', 'de_cache'], [0, 1, 2, 3, 4, 5, 6, 7], inplace=True)
```

```python
df.drop_duplicates(inplace=True)
sum(df.duplicated())
```

```python
x = df.drop(columns=['round_winner'], axis=1)
y = df['round_winner']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

# Modeling

## Logistic Regression

```python
model = LogisticRegression()
model.fit(x_train, y_train)
predictions = model.predict(x_test)
```

```python
print(classification_report(y_test, predictions))
```

```python
tn, fp, fn, tp = confusion_matrix(y_test, predictions).ravel()
print("True Negative: {}".format(tn))
print("False Positive: {}".format(fp))
print("False Negative: {}".format(fn))
print("True Positive: {}".format(tp))
```

```python
accuracy = (tp + tn) / (tp + tn + fp + fn)
specificity = tn / (tn + fp)
precision = tp / (tp + fp)
recall = tp / (tp + fn)
f1 = 2 * ((precision * recall) / (precision + recall))
print("Accuracy: {}".format(accuracy))
print("Specificity: {}".format(specificity))
print("Precision: {}".format(precision))
print("Recall: {}".format(recall))
print("F1 score: {}".format(f1))
```

## Logistic Regression w/ scaled dataset and dropped 'map' feature
